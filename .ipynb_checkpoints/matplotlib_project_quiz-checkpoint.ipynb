## Question 7) The perfect joint (plot)

Let's take a look at this scatterplot. Run the code below (no need to write any code).

```python
sns.lmplot(data=melb_df, x="Price", y="Bedroom2")
```

From this plot we can guess maybe there is some very weak linear relationship? But as it stand right now, it's kind of hard to tell. Why? Because right now our graph is suffering from **overplotting**. Just how many points are there exactly? It's hard to tell since they are so dense they seem to blend together. Lucky for us, seaborn offers some alternative approaches to scatterplots that allow us to see the underlying distrbution of our data better.

Create a joint plot of the dataset with the "Price" and "Bedroom2" variables, with a contour plot as the bivariate visualization and a density curve as the univariate visualization.


```python
# Your code below:

# Answer:
sns.jointplot(data=melb_df, x="Price", y="Bedroom2", kind = 'kde')
```

Now let's look at a hexplot for the data. The bivariate should be a hexplot and the univariate should be a histogram. We are still working with "Price" and "Bedroom2".

```python
# Your code below:

# Answer:
sns.jointplot(data=melb_df, x="Price", y="Bedroom2", kind='hex')
```
